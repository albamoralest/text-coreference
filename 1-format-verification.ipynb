```python
import pandas as pd
```

```python

# Replace 'your_file.csv' with the actual path to your CSV file
csv_file_path = 'filtered_wikidata_dict_ent.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
df['QID'] = df['QID'].astype(str)

# Display the DataFrame
df.info()
```

```python
csv_total = 'TOTAL_download_biographies.csv'
# Read the CSV file into a DataFrame
df_total = pd.read_csv(csv_total)
# Display the DataFrame
df_total.info()
print(df_total)
```

```python
res = []
for item in df.itertuples():
    res_item = []
    # print(item.QID)
    text = str(item.QID)
    # Parse text into a list using the specified separator
    original_list = text.split(';')
    filtered_list = [item_id for item_id in original_list if item_id]
    print(filtered_list)
    res_item.append(filtered_list[0])
    
    for search_text in filtered_list:
        print(search_text)
        try:
            result_df = df_total[df_total['id'] == int(search_text)]
            # matching_rows = df_total[df_total['id'].astype(str).str.contains(search_text, case=False)]
            print(result_df)
            if len(result_df)==0:
                print(f"No match found for text '{search_text}' in the 'id' column.")
                continue
            else:
                # Retrieve the corresponding 'name'
                matched_name = result_df['s'].values[0]
                print('FOUND: ', search_text, matched_name)
                print([filtered_list[0],search_text,matched_name])
                res.append([filtered_list[0],search_text,matched_name])
                break
        except Exception as e: 
            print("ERROR: ", search_text)
            # print(e)
            continue
            
print(res)
```

```python
headers = ['qid', 'dbpedia_id', 'resource']

# Create a DataFrame from the list of lists
df_result = pd.DataFrame(res, columns=headers)
df_result.info()
df_result.head()
```

```python
# Save the DataFrame to a CSV file
df_result.to_csv('matching_bios.csv', index=False)
```
